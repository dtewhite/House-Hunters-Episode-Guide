```python
import pandas as pd
```

```python
base_path = "Resources/House_Hunters_International_Episode_Scrape_"
csv_list = []
for i in range(1,67):
    csv_list.append(f'{base_path + str(i)}.csv')
```

```python
df_list = []
for csv in csv_list:
    df_list.append(pd.read_csv(csv, index_col = 0))
```

```python
combined_df = pd.concat(df_list)
combined_df
```

```python
combined_df.to_csv('Resources/Combined_Episode_Scrapes.csv')
```
